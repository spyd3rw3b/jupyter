# Question 8 (40 points): Data Visualization
## Create a line plot showing the adjusted closing price for `sq`
### Specifications
>- Use `ggplot` for the style
>- Use a `dark_background`
>- Make the figure size have a width of 15 and height of 5
>- Adjust the line width to be 1.5
>- Change the line color to white
>- Name the label 'Square'
>- Show the legend on the chart
>- Show a title on the chart that reads: "Square Stock Data" 
>>- This is a look-up part of the question

```python
style.use('ggplot')

plt.style.use(['dark_background'])

# The following sets the size of the figure.  

mpl.rc('figure', figsize=(15,5))

adjClose.plot(label='Square',color='w')

plt.legend()

plt.title('Square Stock Data')
```
